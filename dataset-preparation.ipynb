```python
import pandas as pd
```

```python
dialogues_df = pd.read_csv('movie_conversations.tsv', 
                 sep="\t", 
                 error_bad_lines=False, 
                 warn_bad_lines=False, 
                 header=None)

dialogues_df.columns=['char_id1', 'char_id2', 'movie_id2', 'frase_ids']
         
dialogues_df.head()
```

```python
characters_df = pd.read_csv('movie_characters_metadata.tsv', 
                            sep='\t', 
                            warn_bad_lines=False,
                            error_bad_lines=False, 
                            header=None)

characters_df.columns=['chId','chName','mId','mName','gender','posCredits']

characters_df.head()
```

```python
lines_df = pd.read_csv('movie_lines.tsv', 
                       sep='\t', 
                       error_bad_lines=False,
                       warn_bad_lines=False, 
                       header=None)

lines_df.columns = ['lineId','chId','mId','chName','dialogue']
lines_df.head()
```

```python
movies_df = pd.read_csv("movie_titles_metadata.tsv", 
                        sep='\t', 
                        error_bad_lines=False,
                        warn_bad_lines=False, 
                        header=None)

movies_df.columns = ['mId','mName','releaseYear','rating','votes','genres']

movies_df.head()
```

```python
df = dialogues_df.merge(characters_df[['chId', 'chName']], 
                       left_on=['char_id1'],
                       right_on=['chId'],
                       how='inner')

df = df.rename(columns={'chName': 'chName_1'})
df = df.drop(columns=['chId'])


df = df.merge(characters_df[['chId', 'chName']], 
              left_on=['char_id2'],
              right_on=['chId'],
              how='inner')

df = df.rename(columns={'chName': 'chName_2'})
df = df.drop(columns=['chId'])

df = df.merge(movies_df[['mId', 'mName', 'genres']], 
              left_on=['movie_id2'],
              right_on=['mId'],
              how='inner')
df = df.drop(columns=['mId'])

df
```

```python
def get_speakers(speaker_1, speaker_2):
    return [speaker_1, speaker_2]
```

```python
df['speakers'] = df.apply(lambda x: get_speakers(x.chName_1, x.chName_2), axis=1)
```

```python
progress_bar = {'L236802': 12.1,
                'L400440': 24.1,
                'L602396': 36.1,
                'L113048': 48.2,
                'L220593': 60.2,
                'L376375': 72.3,
                'L511683': 84.3,
                'L638289': 96.4}
```

```python
def get_replicas(replics_id):
    replics_id = replics_id.replace("'", "").replace("[", "").replace("]", "")
    replicas=[]
    for id in replics_id.split():
        try:
            replicas.append(lines_df[lines_df['lineId']==id]['dialogue'].values[0])
        except:
            replicas.append('Nan')
        if id in progress_bar:
            print(f'{progress_bar[id]}% выполнено')
    return replicas
```

```python
df['replicas'] = df['frase_ids'].apply(lambda x: get_replicas(x))
```

```python
df.to_csv('replicas.csv', index=False)
```
